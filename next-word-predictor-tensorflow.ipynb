## Next word Predictor

```python
file_path = "my_sentences.txt"
```

```python
# Open the file and read its contents
with open(file_path, 'r') as file:
    txt = file.read()
    
print(txt[:500])  # Displaying the first 500 characters to verify the content
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([txt])
```

```python
tokenizer.word_index
```

```python
token_len = len(tokenizer.word_index)   # we need this later, length of vocabulary
print(token_len)
```

```python
# Splitting text into sentences on the basis of '.' or we can use '\n' for upon the basis of newline change
for sentence in txt.split('.'):
    print(sentence)
```

```python
# Tokenized presentation of all the sentences as list of numbers(tokens)
for sentence in txt.split('.'):
   tokenized_sentence = tokenizer.texts_to_sequences([sentence])
   print(tokenized_sentence)
```

```python
input_sequences = []
for sentence in txt.split('.'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python
# we can see the sequence of the words, taking each new words as input from the sentence

input_sequences
```

Here, we can see size of each row is not same, it's not suitable input for the supervised learning, so now we use zero padding fo this

```python
# we need to know the max-length of the sequence
max_length = max([len(x) for x in input_sequences])
print(max_length)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_length, padding = 'pre')

print(padded_input_sequences)
```

```python
X = padded_input_sequences[:,:-1] # we kept all the rows and all columns except the last item in the column

y = padded_input_sequences[:,-1] # we kept all rows but only the last column
```

```python
print(X)
```

```python
X.shape
```

```python
print(y)
```

```python
from tensorflow.keras.utils import to_categorical

y = to_categorical(y, num_classes = token_len+1)

# we did token_len+1 because tokenizer started from 1 instead of zero, where OHE starts from zero
```

```python
print(y)
```

```python
y.shape
```

## Model Training

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(token_len+1, 100,input_length = max_length-1))  
# parameters: size of vocabulary, dimension of embeddings, length of each sentence in input( reduced by 1 as last column is taken as y)
model.add(LSTM(150))   # 150 nodes in LSTm gates
model.add(Dense(token_len+1, activation = 'softmax') )
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y, epochs=25)
```

To predict multiple words

```python
text = 'if you'


for i in range(5):
    #tokenize
    token_text = tokenizer.texts_to_sequences([text])[0]
    #padding
    padded_token_text = pad_sequences([token_text], maxlen = max_length-1 , padding='pre')

    #predict
    pred = model.predict(padded_token_text)

    # finding the position i.e token
    pos = np.argmax(pred)
    print(pos)

    # Decoding token into word
    for word,index in tokenizer.word_index.items():
        if index == pos:
            text = text + " " + word # It adds the o/p again to the i/p and make next prediction
            print(text)
```
